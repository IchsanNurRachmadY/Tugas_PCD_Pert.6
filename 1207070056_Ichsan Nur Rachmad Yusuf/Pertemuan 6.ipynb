Flip Gambar

```python
#Import dependency
import numpy as np                       # Mengimpor library numpy sebagai np
import imageio                           # Mengimpor library imageio
import matplotlib.pyplot as plt          # Mengimpor library matplotlib sebagai plt
```

```python
#Membaca gambar
img = imageio.imread("Gambar1.jpg") #Membaca gambar dan menyimpannya ke variabel img menggunakan modul imageio
```

```python
#Mendapatkan resolusi dan type dari gambar
img_height = img.shape[0] #Mendapatkan tinggi gambar (jumlah piksel dalam dimensi vertikal) dan menyimpannya ke variabel img_height
img_width = img.shape[1] #Mendapatkan lebar gambar (jumlah piksel dalam dimensi horizontal) dan menyimpannya ke variabel img_width
img_channel = img.shape[2] #Mendapatkan jumlah saluran warna pada gambar dan menyimpannya ke variabel img_channel
img_type = img.dtype #Mendapatkan tipe data piksel pada gambar dan menyimpannya ke variabel img_type
```

```python
#Membuat variabel dengan resolusi dan tipe yang sama seperti gambar
img_flip_horizontal = np.zeros(img.shape, img_type) #Membuat array kosong dengan ukuran yang sama dengan gambar dan tipe data yang sama dengan gambar, dan menyimpannya ke variabel img_flip_horizontal
img_flip_vertical = np.zeros(img.shape, img_type) #Membuat array kosong dengan ukuran yang sama dengan gambar dan tipe data yang sama dengan gambar, dan menyimpannya ke variabel img_flip_vertical
```

```python
#Membalik gambar secara horizontal
for y in range(0, img_height): #Looping sebanyak tinggi gambar
    for x in range(0, img_width): #Looping sebanyak lebar gambar
        for c in range(0, img_channel): #Looping sebanyak jumlah saluran warna
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] #Mengisi array img_flip_horizontal dengan piksel gambar yang dicerminkan secara horizontal
```

```python
#Membalik gambar secara vertical
for y in range(0, img_height): #Looping sebanyak tinggi gambar
    for x in range(0, img_width): #Looping sebanyak lebar gambar
        for c in range(0, img_channel): #Looping sebanyak jumlah saluran warna
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] #Mengisi array img_flip_vertical dengan piksel gambar yang dicerminkan secara vertikal
```

```python
#Menampilkan hasil balik gambar
plt.imshow(img_flip_horizontal) # Menampilkan gambar yang telah dicerminkan secara horizontal menggunakan plt.imshow
plt.title("Flip Horizontal") #Menambahkan judul pada gambar yang telah dicerminkan secara horizontal menggunakan plt.title
plt.show() # Menampilkan gambar yang telah dicerminkan secara horizontal
plt.imshow(img_flip_vertical) #Menampilkan gambar yang telah dicerminkan secara vertikal menggunakan plt.imshow
plt.title("Flip Vertical") #Menambahkan judul pada gambar yang telah dicerminkan secara vertikal menggunakan plt.title
plt.show() #Menampilkan gambar yang telah dicerminkan secara vertikal
```
